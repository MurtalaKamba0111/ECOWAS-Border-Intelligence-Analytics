## Objective

This notebook develops a simple interactive Streamlit web application for exploring the simulated ECOWAS trade diversion dataset.

Users can filter the data, visualize key performance indicators, and explore border operations through an interactive web interface.

```python
# ============================================================
# Chapter 2 - Import Libraries
# ============================================================

import pandas as pd

print("Libraries imported successfully.")
```

```python
# ============================================================
# Chapter 3 - Create Streamlit App
# ============================================================

app_code = '''
import streamlit as st
import pandas as pd

st.set_page_config(
    page_title="ECOWAS Trade Dashboard",
    layout="wide"
)

st.title("ECOWAS Border Trade Intelligence Dashboard")

df = pd.read_csv("trade_diversion_transactions.csv")

st.write("Dataset Preview")

st.dataframe(df.head())
'''

with open("app.py", "w") as file:
    file.write(app_code)

print("app.py created successfully.")
```
