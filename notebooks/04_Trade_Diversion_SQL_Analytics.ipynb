## Objective

This notebook demonstrates how SQL can be used to analyze border management data.

Using SQLite within Python, we answer common business questions related to customs revenue, border performance, trade flows, and operational efficiency.

The notebook highlights practical SQL skills frequently required in Data Analyst, Business Intelligence, and Data Science roles.

```python
# ============================================================
# Chapter 1 - Import Libraries
# ============================================================

import pandas as pd
import sqlite3

print("Libraries imported successfully.")
```

```python
# ============================================================
# Chapter 2 - Load Dataset
# ============================================================

df = pd.read_csv("trade_diversion_transactions.csv")

print(df.shape)

df.head()
```

```python
# ============================================================
# Chapter 3 - Create SQLite Database
# ============================================================

connection = sqlite3.connect("trade_diversion.db")

df.to_sql(
    "transactions",
    connection,
    if_exists="replace",
    index=False
)

print("Database created successfully.")
```

## SQL Query 1

Verify that the transaction table has been created successfully.

```python
# ============================================================
# Chapter 4 - Show Tables
# ============================================================

query = """

SELECT name

FROM sqlite_master

WHERE type='table';

"""

tables = pd.read_sql(query, connection)

tables
```

## SQL Query 2

Display the first ten transaction records.

```python
# ============================================================
# Chapter 5 - Sample Records
# ============================================================

query = """

SELECT *

FROM transactions

LIMIT 10;

"""

pd.read_sql(query, connection)
```

## SQL Query 3

Determine the total number of border transactions.

```python
# ============================================================
# Chapter 6 - Total Transactions
# ============================================================

query = """

SELECT COUNT(*) AS Total_Transactions

FROM transactions;

"""

pd.read_sql(query, connection)
```

## SQL Query 4

Identify the busiest border posts based on transaction volume.

```python
# ============================================================
# Chapter 7 - Transactions by Border Post
# ============================================================

query = """

SELECT
    Border_Post,
    COUNT(*) AS Transactions

FROM transactions

GROUP BY Border_Post

ORDER BY Transactions DESC;

"""

pd.read_sql(query, connection)
```

## SQL Query 5

This query identifies the border posts generating the highest customs revenue.

```python
# ============================================================
# Chapter 8 - Revenue by Border Post
# ============================================================

query = """

SELECT
    Border_Post,
    ROUND(SUM(Revenue_Collected_USD),2) AS Total_Revenue

FROM transactions

GROUP BY Border_Post

ORDER BY Total_Revenue DESC;

"""

pd.read_sql(query, connection)
```

## SQL Query 6

This query ranks border posts according to their average queue length.

```python
# ============================================================
# Chapter 9 - Average Queue Length
# ============================================================

query = """

SELECT
    Border_Post,
    ROUND(AVG(Queue_Length),2) AS Average_Queue

FROM transactions

GROUP BY Border_Post

ORDER BY Average_Queue DESC;

"""

pd.read_sql(query, connection)
```

## SQL Query 7

This query identifies border posts with the longest average processing times.

```python
# ============================================================
# Chapter 10 - Average Processing Time
# ============================================================

query = """

SELECT
    Border_Post,
    ROUND(AVG(Processing_Time_Minutes),2) AS Avg_Processing_Time

FROM transactions

GROUP BY Border_Post

ORDER BY Avg_Processing_Time DESC;

"""

pd.read_sql(query, connection)
```

## SQL Query 8

This query ranks commodities by total customs revenue generated.

```python
# ============================================================
# Chapter 11 - Revenue by Commodity
# ============================================================

query = """

SELECT
    Commodity,
    ROUND(SUM(Revenue_Collected_USD),2) AS Total_Revenue

FROM transactions

GROUP BY Commodity

ORDER BY Total_Revenue DESC;

"""

pd.read_sql(query, connection)
```

## SQL Query 9

This query counts high-risk shipments for each commodity.

```python
# ============================================================
# Chapter 12 - High-Risk Shipments
# ============================================================

query = """

SELECT
    Commodity,
    COUNT(*) AS High_Risk_Shipments

FROM transactions

WHERE Smuggling_Risk = 'High'

GROUP BY Commodity

ORDER BY High_Risk_Shipments DESC;

"""

pd.read_sql(query, connection)
```

## SQL Query 10

This query summarizes customs revenue by month to identify revenue trends over time.

```python
# ============================================================
# Chapter 13 - Monthly Revenue Trend
# ============================================================

query = """

SELECT
    strftime('%Y-%m', Transaction_Date) AS Month,
    ROUND(SUM(Revenue_Collected_USD),2) AS Revenue

FROM transactions

GROUP BY Month

ORDER BY Month;

"""

pd.read_sql(query, connection)
```

## SQL Query 11

This query identifies border posts with the highest number of reported security incidents.

```python
# ============================================================
# Chapter 14 - Security Incidents
# ============================================================

query = """

SELECT
    Border_Post,
    COUNT(*) AS Security_Incidents

FROM transactions

WHERE Security_Incident = 'Yes'

GROUP BY Border_Post

ORDER BY Security_Incidents DESC;

"""

pd.read_sql(query, connection)
```

## SQL Query 12

This query uses a SQL window function to rank border posts based on total customs revenue.

```python
# ============================================================
# Chapter 15 - Revenue Ranking
# ============================================================

query = """

SELECT
    Border_Post,
    ROUND(SUM(Revenue_Collected_USD),2) AS Total_Revenue,

    RANK() OVER(
        ORDER BY SUM(Revenue_Collected_USD) DESC
    ) AS Revenue_Rank

FROM transactions

GROUP BY Border_Post;

"""

pd.read_sql(query, connection)
```

## SQL Query 13

This query ranks commodities according to their total trade value.

```python
# ============================================================
# Chapter 16 - Commodity Ranking
# ============================================================

query = """

SELECT
    Commodity,

    ROUND(SUM(Trade_Value_USD),2) AS Trade_Value,

    DENSE_RANK() OVER(
        ORDER BY SUM(Trade_Value_USD) DESC
    ) AS Commodity_Rank

FROM transactions

GROUP BY Commodity;

"""

pd.read_sql(query, connection)
```

## SQL Query 14

This query demonstrates the use of a Common Table Expression (CTE) to identify border posts with above-average customs revenue.

```python
# ============================================================
# Chapter 17 - CTE Example
# ============================================================

query = """

WITH RevenueSummary AS (

    SELECT
        Border_Post,
        SUM(Revenue_Collected_USD) AS Revenue

    FROM transactions

    GROUP BY Border_Post

)

SELECT *

FROM RevenueSummary

WHERE Revenue >

(
    SELECT AVG(Revenue)

    FROM RevenueSummary
);

"""

pd.read_sql(query, connection)
```

## SQL Query 15

A SQL View stores a reusable query for future analysis.

```python
# ============================================================
# Chapter 18 - Create SQL View
# ============================================================

query = """

CREATE VIEW IF NOT EXISTS BorderRevenue AS

SELECT

    Border_Post,

    ROUND(SUM(Revenue_Collected_USD),2) AS Revenue

FROM transactions

GROUP BY Border_Post;

"""

connection.execute(query)

print("SQL View created successfully.")
```

## SQL Query 16

Retrieve data from the SQL View created previously.

```python
# ============================================================
# Chapter 19 - Query SQL View
# ============================================================

query = """

SELECT *

FROM BorderRevenue

ORDER BY Revenue DESC;

"""

pd.read_sql(query, connection)
```

## SQL Query 17

Identify the five border posts with the highest average operational efficiency.

```python
# ============================================================
# Chapter 20 - Efficiency Ranking
# ============================================================

query = """

SELECT

    Border_Post,

    ROUND(AVG(Border_Efficiency_Score),2) AS Avg_Efficiency

FROM transactions

GROUP BY Border_Post

ORDER BY Avg_Efficiency DESC

LIMIT 5;

"""

pd.read_sql(query, connection)
```

## Export SQL Results

Export selected SQL query outputs for reporting and visualization.

```python
# ============================================================
# Chapter 21 - Export SQL Results
# ============================================================

border_revenue = pd.read_sql("""

SELECT
    Border_Post,
    ROUND(SUM(Revenue_Collected_USD),2) AS Revenue

FROM transactions

GROUP BY Border_Post

ORDER BY Revenue DESC;

""", connection)

border_revenue.to_csv(
    "sql_border_revenue.csv",
    index=False
)

print("SQL results exported successfully.")
```
