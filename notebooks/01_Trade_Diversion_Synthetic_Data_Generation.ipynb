# ECOWAS Border Trade Intelligence and Analytics (Version 2)

## Impact Assessment of Regional Trade Diversion Following the AES Countries' Withdrawal from ECOWAS:
### A Data Analytics Case Study of the Tsamiya–Kamba Transit Corridor

---

## Project Background

Regional trade within West Africa depends heavily on efficient cross-border movement of goods, services, and people. Changes in regional political and economic relationships can alter established trade corridors, affecting customs operations, border infrastructure, security, and trade facilitation.

This project investigates a realistic hypothetical scenario in which regional trade routes are diverted following the withdrawal of Burkina Faso, Mali, and Niger (AES countries) from ECOWAS. The analysis focuses on the potential operational impact on the **Tsamiya–Kamba transit corridor** in Kebbi State, Nigeria, which serves as an alternative route connecting Benin and Niger.

---

## Problem Statement

Historically, a significant volume of cargo travelling from the Port of Cotonou (Benin) to Niger passed through the Malanville–Gaya border corridor.

In this hypothetical case study, changes in regional dynamics result in increased use of the Nigerian transit corridor through **Tsamiya** and **Kamba**, leading to:

- Increased border traffic.
- Longer vehicle queues.
- Increased customs workload.
- Changes in customs revenue.
- Increased processing time.
- Higher security risks.
- Greater pressure on existing border infrastructure.

Understanding these potential impacts is essential for strategic planning, efficient resource allocation, and evidence-based policy development.

---

## Project Objectives

This project aims to:

1. Simulate realistic cross-border trade transactions using synthetic data.
2. Measure changes in border traffic before and after the simulated trade diversion.
3. Assess the impact on customs revenue and trade value.
4. Evaluate changes in border processing efficiency.
5. Examine queue length and congestion patterns.
6. Assess changes in security incidents and smuggling risks.
7. Forecast future border traffic using machine learning techniques.
8. Present analytical findings through interactive dashboards and visualizations.

```python
# ============================================================
# ECOWAS VERSION 2
# Trade Diversion Analytics
# Chapter 1 - Import Libraries
# ============================================================

import pandas as pd
import numpy as np
import random

from datetime import datetime, timedelta

random.seed(42)
np.random.seed(42)

print("Libraries imported successfully.")
```

```python
# ============================================================
# Chapter 2 - Project Parameters
# ============================================================

N_RECORDS = 100000

START_DATE = datetime(2022, 1, 1)

END_DATE = datetime(2025, 12, 31)

TOTAL_DAYS = (END_DATE - START_DATE).days

DIVERSION_DATE = datetime(2024, 2, 1)

print("Project parameters created.")
```

```python
# ============================================================
# Chapter 3 - Countries
# ============================================================

COUNTRIES = [

    "Nigeria",
    "Benin",
    "Niger",
    "Ghana",
    "Togo",
    "Burkina Faso",
    "Mali",
    "Cote d'Ivoire",
    "Senegal",
    "Guinea",
    "Liberia",
    "Sierra Leone",
    "Guinea-Bissau",
    "The Gambia",
    "Cape Verde"

]

print("Countries loaded:", len(COUNTRIES))
```

```python
# ============================================================
# Chapter 4 - Border Posts
# ============================================================

BORDER_POSTS = {

    "Nigeria":[
        "Tsamiya",
        "Kamba",
        "Seme",
        "Illela",
        "Jibia"
    ],

    "Benin":[
        "Malanville",
        "Hillacondji"
    ],

    "Niger":[
        "Gaya",
        "Konni"
    ],

    "Ghana":[
        "Aflao",
        "Elubo",
        "Paga"
    ],

    "Togo":[
        "Kodjoviakope"
    ],

    "Burkina Faso":[
        "Cinkanse"
    ],

    "Mali":[
        "Gogui"
    ],

    "Cote d'Ivoire":[
        "Noe",
        "Pogo"
    ],

    "Senegal":[
        "Kidira"
    ],

    "Guinea":[
        "Kouremale"
    ],

    "Liberia":[
        "Bo Waterside"
    ],

    "Sierra Leone":[
        "Gbalamuya"
    ],

    "Guinea-Bissau":[
        "Sao Domingos"
    ],

    "The Gambia":[
        "Amdallai"
    ],

    "Cape Verde":[
        "Praia Port"
    ]

}

print("Border posts loaded.")
```

```python
# ============================================================
# Chapter 5 - Commodities
# ============================================================

COMMODITIES = [

    "Rice",
    "Vehicles",
    "Petroleum",
    "Livestock",
    "Textiles",
    "Electronics",
    "Pharmaceuticals",
    "Construction Materials",
    "Fertilizer",
    "Agricultural Produce"

]

print("Commodity list created.")
```

```python
# ============================================================
# Chapter 6 - Generate Transaction Dates
# ============================================================

transaction_dates = []

for _ in range(N_RECORDS):

    random_days = random.randint(0, TOTAL_DAYS)

    transaction_dates.append(
        START_DATE + timedelta(days=random_days)
    )

print("Transaction dates generated:", len(transaction_dates))
```

```python
# ============================================================
# Chapter 7 - Before / After Diversion
# ============================================================

period = []

for date in transaction_dates:

    if date < DIVERSION_DATE:
        period.append("Before Diversion")
    else:
        period.append("After Diversion")

print(pd.Series(period).value_counts())
```

```python
# ============================================================
# Chapter 8 - Entry Country
# ============================================================

entry_country = []

before_weights = [

    0.12,   # Nigeria
    0.18,   # Benin
    0.12,   # Niger
    0.08,   # Ghana
    0.06,   # Togo
    0.06,   # Burkina Faso
    0.06,   # Mali
    0.06,   # Cote d'Ivoire
    0.05,   # Senegal
    0.05,   # Guinea
    0.04,   # Liberia
    0.04,   # Sierra Leone
    0.03,   # Guinea-Bissau
    0.03,   # The Gambia
    0.02    # Cape Verde

]

after_weights = [

    0.28,   # Nigeria ↑
    0.08,   # Benin ↓
    0.10,
    0.08,
    0.06,
    0.05,
    0.05,
    0.06,
    0.05,
    0.04,
    0.03,
    0.03,
    0.03,
    0.03,
    0.03

]

for p in period:

    if p == "Before Diversion":

        entry_country.append(
            random.choices(
                COUNTRIES,
                weights=before_weights,
                k=1
            )[0]
        )

    else:

        entry_country.append(
            random.choices(
                COUNTRIES,
                weights=after_weights,
                k=1
            )[0]
        )

print(pd.Series(entry_country).value_counts())
```

```python
# ============================================================
# Chapter 9 - Assign Border Posts
# ============================================================

border_post = []

for country, p in zip(entry_country, period):

    if country == "Nigeria":

        if p == "After Diversion":

            border_post.append(
                random.choices(
                    ["Tsamiya", "Kamba", "Seme", "Illela", "Jibia"],
                    weights=[0.40, 0.30, 0.12, 0.10, 0.08],
                    k=1
                )[0]
            )

        else:

            border_post.append(
                random.choices(
                    ["Tsamiya", "Kamba", "Seme", "Illela", "Jibia"],
                    weights=[0.15, 0.15, 0.30, 0.20, 0.20],
                    k=1
                )[0]
            )

    else:

        border_post.append(
            random.choice(BORDER_POSTS[country])
        )

print(pd.Series(border_post).value_counts())
```

```python
# ============================================================
# Chapter 10 - Transaction ID
# ============================================================

transaction_id = []

for i in range(1, N_RECORDS + 1):

    transaction_id.append(f"TRX{i:07d}")

print(transaction_id[:5])
```

```python
# ============================================================
# Chapter 11 - Origin Country
# ============================================================

origin_country = []

origin_weights = [
    0.08,   # Nigeria
    0.30,   # Benin
    0.05,   # Niger
    0.08,   # Ghana
    0.07,   # Togo
    0.05,   # Burkina Faso
    0.04,   # Mali
    0.12,   # Cote d'Ivoire
    0.05,   # Senegal
    0.05,   # Guinea
    0.03,   # Liberia
    0.03,   # Sierra Leone
    0.02,   # Guinea-Bissau
    0.02,   # The Gambia
    0.01    # Cape Verde
]

for _ in range(N_RECORDS):

    origin_country.append(
        random.choices(
            COUNTRIES,
            weights=origin_weights,
            k=1
        )[0]
    )

print(pd.Series(origin_country).value_counts())
```

```python
# ============================================================
# Chapter 12 - Destination Country
# ============================================================

destination_country = []

for p in period:

    if p == "After Diversion":

        destination_country.append(
            random.choices(
                ["Niger","Nigeria","Mali","Burkina Faso"],
                weights=[0.60,0.20,0.10,0.10],
                k=1
            )[0]
        )

    else:

        destination_country.append(
            random.choices(
                ["Nigeria","Niger","Mali","Burkina Faso"],
                weights=[0.40,0.35,0.15,0.10],
                k=1
            )[0]
        )

print(pd.Series(destination_country).value_counts())
```

```python
# ============================================================
# Chapter 13 - Commodity
# ============================================================

commodity = []

before_weights = [
    0.15,
    0.10,
    0.10,
    0.10,
    0.10,
    0.10,
    0.10,
    0.10,
    0.05,
    0.10
]

after_weights = [
    0.18,
    0.08,
    0.18,
    0.08,
    0.08,
    0.10,
    0.08,
    0.12,
    0.05,
    0.05
]

for p in period:

    if p == "Before Diversion":

        commodity.append(
            random.choices(
                COMMODITIES,
                weights=before_weights,
                k=1
            )[0]
        )

    else:

        commodity.append(
            random.choices(
                COMMODITIES,
                weights=after_weights,
                k=1
            )[0]
        )

print(pd.Series(commodity).value_counts())
```

```python
# ============================================================
# Chapter 14 - Vehicle Type
# ============================================================

vehicle_types = [

    "Heavy Truck",
    "Medium Truck",
    "Trailer",
    "Mini Truck",
    "Passenger Bus"

]

vehicle = []

vehicle_weights = [

    0.35,
    0.25,
    0.20,
    0.10,
    0.10

]

for _ in range(N_RECORDS):

    vehicle.append(
        random.choices(
            vehicle_types,
            weights=vehicle_weights,
            k=1
        )[0]
    )

print(pd.Series(vehicle).value_counts())
```

```python
# ============================================================
# Chapter 15 - Trade Value
# ============================================================

trade_value = []

for v in vehicle:

    if v == "Trailer":

        value = np.random.normal(80000,15000)

    elif v == "Heavy Truck":

        value = np.random.normal(65000,12000)

    elif v == "Medium Truck":

        value = np.random.normal(45000,10000)

    elif v == "Mini Truck":

        value = np.random.normal(20000,5000)

    else:

        value = np.random.normal(12000,3000)

    trade_value.append(round(max(1000,value),2))

print("Trade values generated:", len(trade_value))
```

```python
# ============================================================
# Chapter 16 - Cargo Weight (kg)
# ============================================================

cargo_weight = []

for v in vehicle:

    if v == "Trailer":
        weight = np.random.normal(28000, 4000)

    elif v == "Heavy Truck":
        weight = np.random.normal(20000, 3000)

    elif v == "Medium Truck":
        weight = np.random.normal(12000, 2000)

    elif v == "Mini Truck":
        weight = np.random.normal(5000, 1000)

    else:
        weight = np.random.normal(800, 250)

    cargo_weight.append(round(max(100, weight), 2))

print("Cargo weights generated:", len(cargo_weight))
```

```python
# ============================================================
# Chapter 17 - Customs Duty
# ============================================================

customs_duty = []

for value in trade_value:

    rate = random.uniform(0.05, 0.20)

    duty = value * rate

    customs_duty.append(round(duty, 2))

print("Customs duty generated:", len(customs_duty))
```

```python
# ============================================================
# Chapter 18 - Revenue Collected
# ============================================================

revenue = []

for duty in customs_duty:

    revenue_value = duty + np.random.normal(200, 100)

    revenue.append(round(max(0, revenue_value), 2))

print("Revenue generated:", len(revenue))
```

```python
# ============================================================
# Chapter 19 - Queue Length
# ============================================================

queue_length = []

for post, p in zip(border_post, period):

    if p == "After Diversion" and post in ["Tsamiya", "Kamba"]:

        queue = random.randint(40, 120)

    elif post in ["Seme", "Malanville"]:

        queue = random.randint(20, 70)

    else:

        queue = random.randint(5, 40)

    queue_length.append(queue)

print("Queue length generated:", len(queue_length))
```

```python
# ============================================================
# Chapter 20 - Processing Time
# ============================================================

processing_time = []

for queue_size in queue_length:

    time = 15 + (queue_size * 1.3) + np.random.normal(0, 8)

    processing_time.append(round(max(5, time), 1))

print("Processing time generated:", len(processing_time))
```

```python
# ============================================================
# Chapter 21 - Inspection Result
# ============================================================

inspection_result = []

for queue in queue_length:

    if queue > 80:

        result = random.choices(
            ["Passed", "Failed"],
            weights=[0.75, 0.25],
            k=1
        )[0]

    else:

        result = random.choices(
            ["Passed", "Failed"],
            weights=[0.92, 0.08],
            k=1
        )[0]

    inspection_result.append(result)

print(pd.Series(inspection_result).value_counts())
```

```python
# ============================================================
# Chapter 22 - Smuggling Risk
# ============================================================

smuggling_risk = []

for commodity_name, value in zip(commodity, trade_value):

    if commodity_name in ["Petroleum", "Pharmaceuticals", "Electronics"] or value > 70000:

        risk = random.choices(
            ["Low", "Medium", "High"],
            weights=[0.20, 0.40, 0.40],
            k=1
        )[0]

    else:

        risk = random.choices(
            ["Low", "Medium", "High"],
            weights=[0.60, 0.30, 0.10],
            k=1
        )[0]

    smuggling_risk.append(risk)

print(pd.Series(smuggling_risk).value_counts())
```

```python
# ============================================================
# Chapter 23 - Security Incident
# ============================================================

security_incident = []

for risk in smuggling_risk:

    if risk == "High":

        incident = random.choices(
            ["Yes", "No"],
            weights=[0.20, 0.80],
            k=1
        )[0]

    elif risk == "Medium":

        incident = random.choices(
            ["Yes", "No"],
            weights=[0.08, 0.92],
            k=1
        )[0]

    else:

        incident = random.choices(
            ["Yes", "No"],
            weights=[0.02, 0.98],
            k=1
        )[0]

    security_incident.append(incident)

print(pd.Series(security_incident).value_counts())
```

```python
# ============================================================
# Chapter 24 - Border Efficiency Score
# ============================================================

efficiency_score = []

for queue_size, process in zip(queue_length, processing_time):

    score = 100

    score -= queue_size * 0.35

    score -= process * 0.20

    score += np.random.normal(0, 3)

    score = max(20, min(100, round(score, 1)))

    efficiency_score.append(score)

print("Efficiency scores generated:", len(efficiency_score))
```

```python
# ============================================================
# Chapter 25 - Scanner Availability
# ============================================================

scanner_status = []

for post in border_post:

    if post in ["Tsamiya", "Kamba", "Seme", "Malanville"]:

        scanner_status.append(
            random.choices(
                ["Available", "Unavailable"],
                weights=[0.85, 0.15],
                k=1
            )[0]
        )

    else:

        scanner_status.append(
            random.choices(
                ["Available", "Unavailable"],
                weights=[0.60, 0.40],
                k=1
            )[0]
        )

print(pd.Series(scanner_status).value_counts())
```
