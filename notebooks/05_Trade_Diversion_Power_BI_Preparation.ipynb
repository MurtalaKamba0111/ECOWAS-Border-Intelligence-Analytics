## Objective

This notebook prepares analytical datasets for visualization in Microsoft Power BI.

The goal is to transform the transaction-level data into clean summary tables that can be imported directly into Power BI for interactive dashboards and executive reporting.

```python
# ============================================================
# Chapter 1 - Import Libraries
# ============================================================

import pandas as pd

print("Libraries imported successfully.")
```

```python
# ============================================================
# Chapter 2 - Load Dataset
# ============================================================

df = pd.read_csv("trade_diversion_transactions.csv")

print(df.shape)

df.head()
```

## Create Date Dimension

Power BI performs best when using a dedicated Date table for time-based analysis.

```python
# ============================================================
# Chapter 3 - Date Table
# ============================================================

df["Transaction_Date"] = pd.to_datetime(df["Transaction_Date"])
df["Year"] = df["Transaction_Date"].dt.year

df["Month"] = df["Transaction_Date"].dt.month

df["Month_Name"] = df["Transaction_Date"].dt.month_name()

df["Quarter"] = df["Transaction_Date"].dt.quarter
date_table = pd.DataFrame({

    "Date": sorted(df["Transaction_Date"].unique())

})

date_table["Year"] = date_table["Date"].dt.year
date_table["Month"] = date_table["Date"].dt.month
date_table["Month_Name"] = date_table["Date"].dt.month_name()
date_table["Quarter"] = date_table["Date"].dt.quarter

date_table.head()
```

## Border Performance Summary

This table summarizes operational performance by border post.

```python
# ============================================================
# Chapter 4 - Border Summary
# ============================================================

border_summary = (

    df.groupby("Border_Post")

    .agg({

        "Transaction_ID":"count",

        "Trade_Value_USD":"sum",

        "Revenue_Collected_USD":"sum",

        "Queue_Length":"mean",

        "Processing_Time_Minutes":"mean",

        "Border_Efficiency_Score":"mean"

    })

    .reset_index()

)

border_summary
```

## Commodity Summary

Summarize trade activity and revenue by commodity.

```python
# ============================================================
# Chapter 5 - Commodity Summary
# ============================================================

commodity_summary = (

    df.groupby("Commodity")

    .agg({

        "Trade_Value_USD":"sum",

        "Revenue_Collected_USD":"sum",

        "Transaction_ID":"count"

    })

    .reset_index()

)

commodity_summary
```

## Monthly Performance Summary

Prepare monthly performance indicators for trend analysis.

```python
# ============================================================
# Chapter 6 - Monthly Summary
# ============================================================

monthly_summary = (

    df.groupby(["Year","Month_Name"])

    .agg({

        "Trade_Value_USD":"sum",

        "Revenue_Collected_USD":"sum",

        "Transaction_ID":"count"

    })

    .reset_index()

)

monthly_summary
```

## Country Summary

Aggregate transaction activity by entry country.

```python
# ============================================================
# Chapter 7 - Country Summary
# ============================================================

country_summary = (

    df.groupby("Entry_Country")

    .agg({

        "Trade_Value_USD":"sum",

        "Revenue_Collected_USD":"sum",

        "Transaction_ID":"count"

    })

    .reset_index()

)

country_summary
```

## Border Post Dimension

This dimension table contains one record for each border post.

It will be used to create relationships within the Power BI data model.

```python
# ============================================================
# Chapter 8 - Border Dimension
# ============================================================

border_dimension = (

    df[[
        "Border_Post",
        "Entry_Country"
    ]]

    .drop_duplicates()

    .sort_values("Border_Post")

    .reset_index(drop=True)

)

border_dimension.head()
```

## Commodity Dimension

This table contains one record for each commodity.

It simplifies filtering and reporting in Power BI.

```python
# ============================================================
# Chapter 9 - Commodity Dimension
# ============================================================

commodity_dimension = (

    df[["Commodity"]]

    .drop_duplicates()

    .sort_values("Commodity")

    .reset_index(drop=True)

)

commodity_dimension
```

## Country Dimension

This table contains the countries participating in the simulated trade transactions.

```python
# ============================================================
# Chapter 10 - Country Dimension
# ============================================================

country_dimension = (

    df[["Entry_Country"]]

    .drop_duplicates()

    .sort_values("Entry_Country")

    .reset_index(drop=True)

)

country_dimension
```

## Data Quality Assessment

Verify that the dataset contains no missing values before exporting to Power BI.

```python
# ============================================================
# Chapter 11 - Missing Values
# ============================================================

missing = df.isnull().sum()

missing
```

## Dataset Information

Review the structure and data types of the dataset.

```python
# ============================================================
# Chapter 12 - Dataset Information
# ============================================================

df.info()
```

## Export Tables

Export all prepared tables for Power BI.

```python
# ============================================================
# Chapter 13 - Export Tables
# ============================================================

df.to_csv(
    "fact_transactions.csv",
    index=False
)

date_table.to_csv(
    "dim_date.csv",
    index=False
)

border_summary.to_csv(
    "border_summary.csv",
    index=False
)

commodity_summary.to_csv(
    "commodity_summary.csv",
    index=False
)

monthly_summary.to_csv(
    "monthly_summary.csv",
    index=False
)

country_summary.to_csv(
    "country_summary.csv",
    index=False
)

border_dimension.to_csv(
    "dim_border.csv",
    index=False
)

commodity_dimension.to_csv(
    "dim_commodity.csv",
    index=False
)

country_dimension.to_csv(
    "dim_country.csv",
    index=False
)

print("All Power BI tables exported successfully.")
```
